<h2>symbols</h2>

<p>At the core of <code>SymPy</code> is the introduction of symbolic variables that differ quite a bit from <code>Julia</code>'s variables. Symbolic variables do not immediately evaluate to a value, rather the "symbolicness" propagates when interacted with. To keep things manageable, SymPy does some simplifications along the way.</p>

<p>Symbolic expressions are primarily of the <code>Sym</code> type and can be constructed in the standard way:</p>

```julia
x = Sym("x")
```

<p>This creates a symbolic object <code>x</code>, which can be manipulated through further function calls.</p>

<blockquote><p>The @vars macro. For the common case where we wish to create symbolic variables in the <em>main workspace</em>, the <code>@vars</code> macro is useful. It uses the minimal amount of typing to construct new variables. For example, the command <code>@vars a b c</code> creates three variables <code>a</code>, <code>b</code>, and <code>c</code>. The values need not be assigned (and shouldn't be), as the macro automatically does this. (If assigning into the main workspace is not desirable, the related <code>@syms</code> macro is available to reduce typing.) </p></blockquote>

<h3>Assumptions</h3>

<p>With the <code>symbols</code> constructor, it is possible to pass assumptions onto the variables. A list of possible assumptions is <a href="http://docs.sympy.org/dev/modules/core.html#module-sympy.core.assumptions">here</a>. Some examples are:</p>

```julia
x = symbols("x", real=true)
y1, y2 = symbols("y1, y2", positive=true)
alpha = symbols("alpha", integer=true, positive=true)
```

<p>As seen, the <code>symbols</code> function can be used to make one or more variables with one or more assumptions.</p>

<p>We jump ahead for a second to illustrate, but here we see that <code>solve</code> will respect these assumptions, by failing to find solutions to these equations:</p>

```julia
solve(x^2 + 1)   # Â±i are not real
```

```julia
solve(y1 + 1)    # -1 is not positive
```

<h3>Special constants</h3>

<p><code>Julia</code> has its math constants, like <code>pi</code> and <code>e</code>, <code>SymPy</code> as well. A few of these have <code>Julia</code> counterparts provided by <code>SymPy</code>. For example, these three constants are defined (where <code>oo</code> is for infinity):</p>

```julia
PI, E, oo
```

<p>Numeric values themselves can be symbolic. This example shows the difference. The first dispatches to <code>Julia</code>'s <code>asin</code> function, the second to <code>SymPy</code>'s:</p>

```julia
asin(1) , asin(Sym(1))
```

<h2>Substitution</h2>

<p>SymPy provides a means to substitute values in for the symbolic expressions. The specification requires an expression, a variable in the expression to substitute in for, and a new value. For example, this is one way to make a polynomial in a new variable:</p>

```julia
x, y = symbols("x,y")
ex = x^2 + 2x + 1
subs(ex, x, y)
```

<p>Substitution can also be numeric:</p>

```julia
subs(ex, x, 0)
```

<p>The output has no free variables, but is still symbolic.</p>

<p>Expressions with more than one variable can have multiple substitutions, where each is expressed as a tuple:</p>

```julia
x,y,z = symbols("x,y,z")
ex = x + y + z
subs(ex, (x,1), (y,pi))
```

<p>When used with the pipeline operator, <code>|></code>, there is a curried form that allows the expression to be implicit:</p>

```julia
ex |> subs(x, 1)
```

<p>As <code>subs</code> is very similar in spirit to <code>Julia</code>'s <code>replace</code> function, an alias is provided:</p>

```julia
ex |> replace(y, pi)
```

<h2>Conversion from symbolic to numeric</h2>

<p>SymPy provides two identical means to convert a symbolic math expression to a number. One is <code>evalf</code>, the other <code>N</code>. Within <code>Julia</code> we decouple this, using <code>N</code> to also convert to a <code>Julian</code> value and <code>evalf</code> to leave the conversion as a symbolic object.  The <code>N</code> function converts symbolic integers, rationals, irrationals, and complex values, while attempting to find an appropriate <code>Julia</code> type for the value.</p>

<p>To see the difference, we use both on <code>PI</code>:</p>

```julia
N(PI)  # floating-point value
```

<p>Whereas, this may look the same, it is still symbolic:</p>

```julia
evalf(PI)
```

<p>Both allow for a precision argument to be passed through the second argument. This is how 30 digits of $\pi$ can be extracted:</p>

```julia
N(PI, 30)
```

<p>Here <code>N</code> produces a <code>BigFloat</code> with a precision to match (basically) the specified number of digits. Whereas</p>

```julia
evalf(PI, 30)
```

<p>leaves the value as a symbolic object with 30 digits of accuracy.</p>

<h2>algebraic expressions</h2>

<p><code>SymPy</code> overloads many of <code>Julia</code>'s functions to work with symbolic objects, such as seen above with <code>asin</code>. The usual mathematical operations such as <code>+</code>, <code>*</code>, <code>-</code>, <code>/</code> etc. work through <code>Julia</code>'s promotion mechanism, where numbers are promoted to symbolic objects, others dispatch internally to related <code>SymPy</code> functions.</p>

<p>In most all  cases, thinking about this distinction between numbers and symbolic numbers is unnecessary, as numeric values passed to <code>SymPy</code> functions are typically promoted to symbolic expressions. This conversion will take math constants to their corresponding <code>SymPy</code> counterpart, rational expressions to rational expressions, and floating point values to floating point values. However there are edge cases. An expression like <code>1//2 * pi * x</code> will differ from the seemingly identical  <code>1//2 * (pi * x)</code>. The former will produce a floating point value from <code>1//2 * pi</code> before being promoted to a symbolic instance. Using the symbolic value <code>PI</code> makes this expression work either way.</p>

<p>Most of <code>Julia</code>'s <a href="http://julia.readthedocs.org/en/latest/manual/mathematical-operations/#elementary-functions">mathematical</a> functions are overloaded to work with symbolic expressions. <code>Julia</code>'s generic definitions are used, as possible. This also introduces some edge cases. For example, <code>x^(-2)</code> will balk due to the negative, integer exponent, but either <code>x^(-2//1)</code> or <code>x^Sym(-2)</code> will work as expected, as the former call first dispatches to a generic defintion, but the latter expressions two do not.</p>

<p><code>SymPy</code> makes it very easy to work with polynomial and rational expressions. First we create some variables:</p>

```julia
x,y,z = symbols("x, y, z")
```

<h3>expand, factor, collect, simplify</h3>

<p>A typical polynomial expression in a single variable can be written in two common ways, expanded or factored form. Using <code>factor</code> and <code>expand</code> can move between the two.</p>

<p>For example,</p>

```julia
p = x^2 + 3x + 2
factor(p)
```

<p>Or</p>

```julia
expand(prod([(x-i) for i in 1:5]))
```

<p>The <code>factor</code> function factors over the rational numbers, so something like this with obvious factors is not finished:</p>

```julia
factor(x^2 - 2)
```

<p>When expressions involve one or more variables, it can be convenient to be able to manipulate them. For example, if we define <code>q</code> by:</p>

```julia
q = x*y + x*y^2 + x^2*y + x
```

<p>Then we can collect the terms by the variable <code>x</code>:</p>

```julia
collect(q, x)
```

<p>or the variable <code>y</code>:</p>

```julia
collect(q, y)
```

<p>These are identical expressions, though viewed differently. </p>

<p>A more broad-brush approach is to let <code>SymPy</code> simplify the values. In this case, the common value of <code>x</code> is factored out:</p>

```julia
simplify(q)
```

<p>The <code>simplify</code> function attempts to apply the dozens of functions related to simplification that are part of SymPy. It is also possible to apply these functions one at a time, for example <code>trigsimp</code> does trigonometric simplifications.</p>

<p>The SymPy tutorial illustrates that <code>expand</code> can also result in simplifications through this example:</p>

```julia
expand((x + 1)*(x - 2) - (x - 1)*x)
```

<p>These methods are not restricted to polynomial expressions and will work with other expressions. For example, <code>factor</code> identifies the following as a factorable object in terms of the variable <code>exp(x)</code>:</p>

```julia
factor(exp(2x) + 3exp(x) + 2)
```

<h2>Rational expressions: apart, together, cancel</h2>

<p>When working with rational expressions, SymPy does not do much simplification unless asked. For example:</p>

```julia
r = 1/x + 1/x^2
```

<p>is not simplified. To put over a common denominator, the <code>together</code> function is available:</p>

```julia
together(r)
```

<p>The <code>apart</code> function does the reverse, creating a partial fraction decomposition:</p>

```julia
apart( (4x^3 + 21x^2 + 10x + 12) /  (x^4 + 5x^3 + 5x^2 + 4x))
```

<p>Some times factors are canceled, as here:</p>

```julia
top = (x-1)*(x-2)*(x-3)
bottom = (x-1)*(x-4)
top/bottom
```

<p>(This might make math faculty a bit upset, but it is in line with student thinking.)</p>

<p>However, expanded terms are not factored, then canceled:</p>

```julia
r = expand(top) / expand(bottom)
```

<p>The <code>cancel</code> function will do so. It takes rational functions and puts them in a canonical $p/q$ form with no common (rational) factors and leading terms which are integers:</p>

```julia
cancel(r)
```

<h2>Powers</h2>

<p>The SymPy <a href="http://docs.sympy.org/dev/tutorial/simplification.html#powers">tutorial</a> offers a thorough explanation on powers and which get simplified and under what conditions. Basically</p>

<ul><li>$x^a x^b = x^{a+b}$ is always true. However</li></ul>

<ul><li>$x^a y^a=(xy)^a$ is only true with assumptions, such as $x,y \geq 0$ and $a$ is real, but not in general. For example, $x=y=-1$ and $a=1/2$ has $x^a \cdot y^a = i \cdot i =  -1$, where as $(xy)^a = 1$.</li></ul>

<ul><li>$(x^a)^b = x^{ab}$ is only true with assumptions. For example $x=-1, a=2$, and $b=1/2$ gives $(x^a)^b = 1^{1/2} = 1$, whereas $x^{ab} = -1^1 = -1$.</li></ul>

<p>We see that with assumptions, the expression does simplify to $0$:</p>

```julia
x,y = symbols("x,y", nonnegative=true)
a = symbols("a", real=true)
simplify(x^a * y^a - (x*y)^a)
```

<p>However, without assumptions this is not the case</p>

```julia
x,y,a = symbols("x,y,a")
simplify(x^a * y^a - (x*y)^a)
```

<p>The <code>simplify</code> function calls <code>powsimp</code> above. The <code>powsimp</code> function has the keyword argument <code>force=true</code> to force simplification even if assumptions are not specified:</p>

```julia
powsimp(x^a * y^a - (x*y)^a, force=true)
```

<h2>Trigonometric simplification</h2>

<p>For trigonometric expressions, <code>simplify</code> will use <code>trigsimp</code> to simplify:</p>

```julia
theta = symbols("theta", real=true)
p = cos(theta)^2 + sin(theta)^2
```

<p>Calling either <code>simplify</code> or <code>trigsimp</code> will apply the Pythagorean identity:</p>

```julia
simplify(p)
```

<p>The <code>trigsimp</code> formula is, of course, course aware of the double angle formulas:</p>

```julia
simplify(sin(2theta) - 2sin(theta)*cos(theta))
```

<p>The <code>expand_trig</code> function will expand such expressions:</p>

```julia
expand_trig(sin(2theta))
```

<h2>Coefficients</h2>

<p>Returning to polynomials, there are a few functions to find various pieces of the polynomials. First we make a general quadratic polynomial:</p>

```julia
a,b,c,x = symbols("a, b, c, x") 
p = a*x^2 + b*x + c
```

<p>The <code>coeff(ex, monom)</code> function will return the corresponding coefficient of the monomial:</p>

```julia
coeff(p, x^2) # a 
coeff(p, x)   # b
```

<p>The constant can be found through substitution:</p>

```julia
subs(p, x, 0) # c
```

<p>Though one could use some trick like this to find all the coefficients:</p>

```julia
[[coeff(p, x^i) for i in N(degree(p)):-1:1], subs(p,x,0)]
```

<p>that is cumbersome, at best. SymPy has a function <code>coeffs</code>, but it is defined for polynomial types, so will fail on <code>p</code>:</p>

```julia
coeffs(p) # fails
```

<p>Polynomials are a special class in SymPy and must be constructed. The <code>Poly</code> constructor can be used. As there is more than one free variable in <code>p</code>, we specify the variable <code>x</code> below:</p>

```julia
q = Poly(p, x)
coeffs(q)
```

<h2>Polynomial roots: solve, real_roots, polyroots, nroots</h2>

<p>SymPy provides functions to find the roots of a polynomial. In general, a polynomial with real coefficients of degree $n$ will have $n$ roots when multiplicities and complex roots are accounted for. The number or real roots is consequently between $0$ and $n$.</p>

<p>For a <em>univariate</em> polynomial expression (a single variable), the real roots, when available, are returned by <code>real_roots</code>. For example,</p>

```julia
real_roots(x^2 - 2)
```

<p>Unlike <code>factor</code> â which only factors over rational factors â <code>real_roots</code> finds the two irrational roots here. It is well known (the <a href="http://en.wikipedia.org/wiki/Abel%E2%80%93Ruffini_theorem">Abel-Ruffini theorem</a>) that for degree 5 polynomials, or higher, it is not always possible to express the roots in terms of radicals. However, when the roots are rational <code>SymPy</code> can have success:</p>

```julia
p = (x-3)^2*(x-2)*(x-1)*x*(x+1)*(x^2 + x + 1)
real_roots(p)
```

<p>In this example, the degree of <code>p</code> is 8, but there are only 6 real roots returned, the double root of $3$ is accounted for, but the two complex roots of <code>x^2 + x+ 1</code> are not. The actual roots can be found with <code>solve</code>:</p>

```julia
solve(p)
```

<p>This finds the complex roots, but does not account for the double root. The <code>roots</code> function of SymPy does. This particular function is not exported (as it conflicts with the <code>roots</code> function from <code>Polynomials</code> and <code>Roots</code>) but we can still access it using indexing notation or its alias <code>polyroots</code>.</p>

<blockquote><p>Indexing with a symbol. When a symbolic expression is indexed by a symbol it returns a function which maps to a corresponding SymPy function. For example, <code>p[:roots](args...)</code> will call <code>roots(p, args...)</code> within SymPy. For methods of SymPy objects, the same is true, so if <code>roots</code> were a class method, then the call would resolve to <code>p.roots(args...)</code>. </p></blockquote>

<p>The output of calling <code>roots</code> will be a dictionary whose keys are the roots and values the multiplicity.</p>

```julia
p[:roots]()     # or polyroots(p)
```

<p>When exact answers are not provided, the <code>:roots</code> call is contentless:</p>

```julia
p = x^5 - x + 1
polyroots(p)
```

<p>Calling <code>solve</code> seems to produce very little as well:</p>

```julia
rts = solve(p)
```

<p>But in fact, <code>rts</code> contains lots of information. We can extract numeric values quite easily with <code>N</code>:</p>

```julia
[N(r) for r in rts]     # or map(N, rts)
```

<p>These are numeric approximations to irrational numbers. For such a task, the <code>nroots</code> function will also work, though with this the answers are still symbolic:</p>

```julia
nroots(p)
```

<h2>solve</h2>

<p>The <code>solve</code> function is more general purpose than just finding roots of univariate polynomials. The function tries to solve for when an expression is 0, or a set of expressions are all 0.</p>

<p>For example, it can be used to solve when $\cos(x) = \sin(x)$:</p>

```julia
solve(cos(x) - sin(x))
```

<p>Though there are infinitely many correct solutions, these are within a certain range.</p>

<p>The <code>solve</code> function has limits. For example, there is no symbolic solution here:</p>

```julia
solve(cos(x) - x)
```

<p>For such, a numeric method would be needed.</p>

<p>Though it can't solve everything, the <code>solve</code> function can also solve equations of a more general type. For example, here it is used to derive the quadratic equation:</p>

```julia
a,b,c  = symbols("a,b,c", real=true)
p = a*x^2 + b*x + c
solve(p, x)
```

<p>The extra argument <code>x</code> is passed to <code>solve</code> so that <code>solve</code> knows which variable to solve for. If not given, <code>solve</code> tries to find a solution with all the free variables, which in this case is not helpful:</p>

```julia
solve(p)
```

<p>Systems of equations can be solved as well. We specify them within a vector of expressions, <code>[ex1, ex2, ..., exn]</code> where a found solution is one where all the expressions are 0. For example, to solve this linear system: $2x + 3y = 6, 3x - 4y=12$, we have:</p>

```julia
x, y = symbols("x,y", real=true)
exs = [2x+3y-6, 3x-4y-12]
d = solve(exs)
```

<p>Plugging the solutions into the equations to check is a bit cumbersome, as the keys of the dictionary that is returned are strings, but functions like <code>subs</code> prefer the variables. Here is one workaround:</p>

```julia
vars = [x,y]
vals = [d[string(var)] for var in vars]
[subs(ex, zip(vars, vals)...) for ex in exs]
```

<p>In the previous example, the system had two equations and two unknowns. When that is not the case, one can specify the variables to solve for as a vector. In this example, we find a quadratic polynomial that approximates $\cos(x)$ near $0$:</p>

```julia
a,b,c,h = symbols("a,b,c,h", real=true)
p = a*x^2 + b*x + c
fn = cos
exs = [fn(0*h)-subs(p,x,0), fn(h)-subs(p,x,h), fn(2h)-subs(p,x,2h)]
d = solve(exs, [a,b,c])
```

<p>Again, a dictionary is returned. The polynomial itself can be found by substituting back in for <code>a</code>, <code>b</code>, and <code>c</code>:</p>

```julia
vars=[a,b,c]
vals = [d[string(var)] for var in vars]
quad_approx = subs(p, zip(vars, vals)...)
```

<p>(Taking the limit as $h$ goes to 0 produces the answer $1 - x^2/2$.)</p>

<h3>Solving using logical operators</h3>

<p>The <code>solve</code> function does not need to just solve <code>ex = 0</code>. There are other means to specify an equation. Ideally, it would be nice to say <code>ex1 == ex2</code>, but the interpretation of <code>==</code> is not for this. Rather, <code>SymPy</code> introduces <code>Eq</code> for equality:</p>

```julia
solve(Eq(x, 1))
```

<p>Gives 1, as expected from solving <code>x == 1</code>. In addition to <code>Eq</code>, there are <code>Lt</code>, <code>Le</code>, <code>Ge</code>, <code>Gt</code>. The Unicode operators are not aliased to these, but there are alternatives <code>\ll[tab]</code>, <code>\le[tab]</code>, <code>\Equal[tab]</code>, <code>\ge[tab]</code>, <code>\gg[tab]</code> and <code>\neg[tab]</code> to negate.</p>

<p>So, the above could have been written with the following nearly identical expression, though it is entered with <code>\Equal[tab]</code>. </p>

```julia
solve(x â©µ 1)
```

<p>Here is an alternative way of asking a previous question on a pair of linear equations:</p>

```julia
x, y = symbols("x,y", real=true)
exs = [2x+3y â©µ 6, 3x-4y â©µ 12]    ## Using \Equal[tab]
d = solve(exs)
```

<h2>Calculus</h2>

<p><code>SymPy</code> has many of the basic operations of calculus provided through a handful of functions.</p>

<h3>limits</h3>

<p>Limits are computed by the <code>limit</code> function which takes an expression, a variable and a value, and optionally a direction specified by either <code>dir="+"</code> or <code>dir="-"</code>.</p>

<p>For example, this shows Gauss was right:</p>

```julia
limit(sin(x)/x, x, 0)
```

<p>Limits at infinity are done by using <code>oo</code> for $\infty$:</p>

```julia
limit((1+1/x)^x, x, oo)
```

<p>This example computes what L'Hopital reportedly paid a Bernoulli for</p>

```julia
a = symbols("a", positive=true)
ex = (sqrt(2a^3*x-x^4) - a*(a^2*x)^(1//3)) / (a - (a*x^3)^(1//4))
```

<p>The limit as $x$ goes to $a$ gives an indeterminate expression:</p>

```julia
subs(ex, x, a)
```

<p>we can see it is of the form $0/0$:</p>

```julia
subs(denom(ex), x, a), subs(numer(ex), x, a)
```

<p>And we get</p>

```julia
limit(ex, x, a)
```

<p>In a previous example, we defined <code>quad_approx</code>. The limit as <code>h</code> goes to $0$ gives <code>1 - x^2/2</code>, as expected:</p>

```julia
limit(quad_approx, h, 0)
```

<h4>left and right limits</h4>

<p>The limit is defined when both the left and right limits exist and are equal. But left and right limits can exist and not be equal. The <code>sign</code> function is $1$ for positive $x$, $-1$ for negative $x$ and $0$ when $x$ is 0. It should not have a limit at $0$:</p>

```julia
limit(sign(x), x, 0)
```

<p>Oops. Well, the left and right limits are different anyways:</p>

```julia
limit(sign(x), x, 0, dir="-"), limit(sign(x), x, 0, dir="+")
```

<h4>operator interface</h4>

<p>For univariate functions there is an "operator" interface, where we pass a function object as the first argument and the value for <code>c</code> as the second (the variable is implicit, as <code>f</code> has only one).</p>

```julia
f(x) = sin(5x)/x
limit(f, 0)
```

<h4>numeric limits</h4>

<p>The <code>limit</code> function uses the <a href="http://docs.sympy.org/latest/modules/series.html#the-gruntz-algorithm">Gruntz</a> algorithm. It is far more reliable then simple numeric attempts at integration. An example of Gruntz is the right limit at $0$ of the function:</p>

```julia
f(x) = 1/x^(log(log(log(log(1/x)))) - 1)
```

<p>A numeric attempt might be done along these lines:</p>

```julia
hs = [10.0^(-i) for i in 6:16]
ys = [f(h) for h in hs]
[hs ys]
```

<p>With a values appearing to approach $0$. However, in fact these values will ultimately head  off to $\infty$:</p>

```julia
limit(f(x), x, 0, dir="+")
```

<h3>derivatives</h3>

<p>One <em>could</em> use limits to implement the definition of a derivative:</p>

```julia
x, h = symbols("x,h")
f(x) = exp(x)*sin(x)
limit((f(x+h) - f(x)) / h, h, 0)
```

<p>However, it would be pretty inefficient, as <code>SymPy</code> already does a great job with derivatives. The <code>diff</code> function implements this. The basic syntax is <code>diff(ex, x)</code> to find the first derivative in <code>x</code> of the expression in <code>ex</code>, or its generalization to $k$th derivatives with <code>diff(ex, x, k)</code>.</p>

<p>The same derivative computed above by a limit could be found with:</p>

```julia
diff(f(x), x)
```

<p>Similarly, we can compute other derivatives:</p>

```julia
diff(x^x, x)
```

<p>Or</p>

```julia
diff(exp(-x^2), x, 2)
```

<p>The output is a simple expression, so can be composed with other functions, such as <code>solve</code>. For example, here we find the critical points of a rational function:</p>

```julia
f(x) = (12x^2 - 1) / (x^3)
diff(f(x), x) |> solve
```

<h4>Operator version</h4>

<p><code>SymPy</code> provides an "operator" version of <code>diff</code> for univariate functions for convenience (<code>diff(f::Function,k=1)=diff(f(x),x,k)</code>):</p>

```julia
f(x) = exp(x)*cos(x)
diff(f, 2)
```

<h4>Partial derivatives</h4>

<p>The <code>diff</code> function makes finding partial derivatives as easy as specifying the variable to differentiate in. This  example computes the mixed partials of an expression in <code>x</code> and <code>y</code>:</p>

```julia
x,y = symbols("x,y")
ex = x^2*cos(y)
Sym[diff(diff(ex,v1), v2) for v1 in [x,y], v2 in [x,y]]
```

<p>The key is the iterated use of <code>diff</code>. The extra <code>Sym</code>, of the form <code>T[]</code>, helps <code>Julia</code> resolve the type of the output.</p>

<h4>Unevaluated derivatives</h4>

<p>The <code>Derivative</code> function provides unevaluated derivatives, useful with differential equations and the output for unknown functions. Here is an example:</p>

```julia
ex = Derivative(exp(x*y), x, y, 2)
```

<p>(The <code>y,2</code> is a replacement for <code>y,y</code> which makes higher order terms easier to type.) These expressions are evaluated with <code>doit</code>:</p>

```julia
doit(ex)
```

<h4>Implicit derivatives</h4>

<p>SymPy can be used to find derivatives of implicitly defined functions. For example, the task of finding $dy/dx$ for the equation:</p>

<p>$$ y^4 - x^4 -y^2 + 2x^2 = 0 $$</p>

<p>As with the mathematical solution, the key is to treat one of the variables as depending on the other. In this case, we think of $y$ locally as a function of $x$. SymPy allows us to create symbolic functions, and we will use one to substitute in for <code>y</code>.</p>

<p>In SymPy, symbolic functions use the class name  "Function", but in <code>SymPy</code> we use <code>SymFunction</code> to avoid a name collision with one of <code>Julia</code>'s primary types. The constructor can be used as <code>SymFunction(:F)</code>, or more clearly passed to the <code>symbols</code> function through the <code>cls</code> keyword:</p>

```julia
F, G = symbols("F, G", cls=SymFunction)
```

<p>We can call these functions, but we get a function expression:</p>

```julia
F(x)
```

<p>SymPy can differentiate symbolically, again with <code>diff</code>:</p>

```julia
diff(F(x))
```

<p>To get back to our problem, we have our expression:</p>

```julia
x,y = symbols("x, y")
ex = y^4 - x^4 - y^2 + 2x^2
```

<p>Now we substitute:</p>

```julia
ex1 = subs(ex, y, F(x))
```

<p>We want to differentiate "both" sides. As the right side is just $0$, there isn't anything to do here, but mentally keep track. As for the left we have:</p>

```julia
ex2 = diff(ex1, x)
```

<p>Now we collect terms and solve in terms of $F'(x)$</p>

```julia
ex3 = solve(ex2, diff(F(x)))[1]
```

<p>Finally, we substitute back into the solution for $F(x)$:</p>

```julia
ex4 = subs(ex3, F(x), y)
```

<p>To visualize this, SymPy's <code>plot_implicit</code> function could be used. This requires <code>PyPlot</code> to be the graphing package, but we are using <code>Gadfly</code>. (The <code>PyPlot</code> call is just `plot_implicit(Eq(ex, 0))<code>). We use the `ImplicitEquations</code> package which provides a similar means to plot implicit equations within <code>Gadfly</code>, though we need to create functions and not use our symbolic expressions.</p>

```julia
using ImplicitEquations

f(x,y) = y^4 - x^4 - y^2 + 2x^2

x0, y0 = (0,1)
m = N(subs(ex3, (x,x0), (y,y0)))
g(x,y) = y - (y0 + m*(x-x0))

ImplicitEquations.ggraph((f==0) | (g==0))  # ugly default colors...
```

<h6>Example: A Norman Window</h6>

<p>A classic calculus problem is to maximize the area of a <a href="http://en.wiktionary.org/wiki/Norman_window">Norman window</a> (in the shape of a rectangle with a half circle atop) when the perimeter is fixed to be $P \geq 0$.</p>

<p>Label the rectangle as $w$ and $h$ and then the half circle has radius $r=w/2$. With this, we can see that the area is $wh+(1/2)\pi r^2$ and the perimeter is $w + 2h + \pi r$. This gives:</p>

```julia
w, h, P = symbols("w, h, P", nonnegative=true)
r = w/2
A = w*h + 1//2 * (pi * r^2)
p = w + 2h + pi*r
```

<p>(There is a subtlety above, as <code>1//2*pi*r^2</code> will not work as expected, as the products will be done left to right, and <code>1//2*pi</code> will be converted to floating point before multiplying <code>r^2</code>, as such we rewrite the terms. It may be easier to use <code>PI</code> instead of <code>pi</code>.)</p>

<p>We want to solve for <code>h</code> from when <code>p=P</code> (our fixed value) and substitute back into <code>A</code>. We solve <code>P-p==0</code>:</p>

```julia
A1 = subs(A, h, solve(P-p, h)[1])
```

<p>Now we note this is a parabola in <code>w</code>, so any maximum will be an endpoint or the vertex, provided the leading term is negative. To see the leading term we could do this:</p>

```julia
coeffs(Poly(A1, w))
```

<p>Or without using the <code>Poly</code> methods, we could do this:</p>

```julia
coeff(collect(expand(A1), w), w^2)
```

<p>Either way, we see the leading coefficient, $-1/2 - \pi/8$, is negative, so the maximum can only happen at an endpoint or the vertex of the parabola. Now we check that when $w=0$ (the left endpoint) the area is $0$:</p>

```julia
subs(A1, w, 0)
```

<p>The other endpoint is when $h=0$, or</p>

```julia
b = solve(subs(P-p, h, 0), w)[1]
```

<p>We will need to check the area at <code>b</code> and at the vertex.</p>

<p>To find the vertex, we can use calculus â it will be when the derivative in <code>w</code> is $0$:</p>

```julia
c = solve(diff(A1, w), w)[1]
```

<p>The answer will be the larger of <code>A1</code> at <code>b</code> or <code>c</code>:</p>

```julia
atb = subs(A1, w, b)
atc = subs(A1, w, c)
```

<p>A simple comparison isn't revealing:</p>

```julia
atc - atb
```

<p>But after simplifying, we can see that this expression is positive if $P$ is:</p>

```julia
simplify(atc - atb)
```

<p>With this, we see that the maximum area happens at <code>c</code> with area <code>atc</code>.</p>

<h3>integrals</h3>

<p>Integration is implemented in SymPy through the <code>integrate</code> function. There are two basic calls:  <code>integrate(f(x), x)</code> will find the indefinite integral ($\int f(x) dx$) and when endpoints are specified through <code>integrate(f(x), (x, a, b))</code> the definite integral will be found ($\int_a^b f(x) dx$). The special form <code>integrate(ex, x, a, b)</code> can be used for single integrals, but the specification through a tuple is needed for multiple integrals.</p>

<p>Basic integrals are implemented:</p>

```julia
integrate(x^3, x)
```

<p>Or in more generality:</p>

```julia
n = symbols("n", real=true)
ex = integrate(x^n, x)
```

<p>The output here is a <em>piecewise function</em>, performing a substitution will choose a branch in this case:</p>

```julia
subs(ex, n, 3)
```

<p>Definite integrals are just as easy. Here is Archimedes' answer:</p>

```julia
integrate(x^2, (x, 0, 1))
```

<p>Tedious problems, such as those needing multiple integration-by-parts steps can be done easily:</p>

```julia
integrate(x^5*sin(x), x)
```

<p>The SymPy tutorial says:</p>

<p>"<code>integrate</code> uses powerful algorithms that are always improving to compute both definite and indefinite integrals, including heuristic pattern matching type algorithms, a partial implementation of the Risch algorithm, and an algorithm using Meijer G-functions that is useful for computing integrals in terms of special functions, especially definite integrals."</p>

<p>The tutorial gives the following example:</p>

```julia
f(x) = (x^4 + x^2 * exp(x) - x^2 - 2x*exp(x) - 2x - exp(x)) * exp(x) / ( (x-1)^2 * (x+1)^2 * (exp(x) + 1) )
integrate(f(x), x)
```

<h4>Multiple integrals</h4>

<p>The <code>integrate</code> function uses a tuple, <code>(var, a, b)</code>, to specify the limits of a definite integral. This syntax lends itself readily to multiple integration.</p>

<p>For example, the following computes the integral of $xy$ over the unit square:</p>

```julia
x, y = symbols("x,y")
integrate(x*y, (y, 0, 1), (x, 0, 1))
```

<p>The innermost terms can depend on outer ones. For example, the following integrates $x^2y$ over the upper half of the unit circle:</p>

```julia
integrate(x^2*y, (y, 0, sqrt(1 - x^2)), (x, -1, 1))
```

<h4>Unevaluated integrals</h4>

<p>The <code>Integral</code> function can stage unevaluated integrals that will be evaluated by calling <code>doit</code>. It is also used when the output is unknown. This example comes from the tutorial:</p>

```julia
integ = Integral(sin(x^2), x)
```

```julia
doit(integ)
```

<h4>operator version</h4>

<p>For convenience, for univariate functions there is a convenience wrapper so that the operator styles â <code>integrate(f)</code> and <code>integrate(f, a, b)</code> â will perform the integrations.</p>

```julia
f(x) = exp(x) * cos(x)
integrate(f)
```

<h3>Taylor series</h3>

<p>The <code>series</code> function can compute series expansions around a point to a specified order. For example,</p>

```julia
removeO(s1)
```

<h2>differential equations</h2>

<p>SymPy has facilities for solving ordinary differential <a href="http://docs.sympy.org/latest/modules/solvers/ode.html">equations</a>. The key is to create a symbolic function expression using the <code>SymFunction</code> class. Again, this is done through:</p>

```julia
F = symbols("F", cls=SymFunction)
```

<p>With this, we can  construct a  differential equation. Following the SymPy tutorial, we solve $f''(x) - 2f'(x) + f(x) = \sin(x)$:</p>

```julia
diffeq = Eq(diff(F(x), x, 2) - 2*diff(F(x)) + F(x), sin(x))
```

<p>(We could have used the Unicode <code>\Equal[tab]</code> to make this look even closer to the math equation.)</p>

<p>With this, we just need the <code>dsolve</code> function. This is called as <code>dsolve(eq, variable)</code>:</p>

```julia
ex = dsolve(diffeq, F(x))
```

<p>This solution has two constants, $C_1$ and $C_2$, that would be found from initial conditions. Say we know $F(0)=0$ and $F'(0)=1$, can we find the constants? To work with the returned expression, it is most convenient to get just the right hand side. The <code>rhs</code> function will return that:</p>

```julia
ex1 = rhs(ex)
```

<p>(The <a href="http://docs.sympy.org/dev/modules/core.html#sympy.core.basic.Basic.args">args</a> function also can be used to break up the expression into parts.)</p>

<p>With this, we can solve for <code>C1</code> from substituting in $0$ for $x$:</p>

```julia
solve(subs(ex1, x, 0), Sym("C1"))
```

<p>We see that $C1=-1/2$, which we substitute in:</p>

```julia
ex2 = subs(ex1, Sym("C1"), -1//2)
```

<p>We know that $F'(0)=1$ now, so we solve for <code>C2</code> through</p>

```julia
solve( subs(diff(ex2, x), x, 0) - 1, Sym("C2") )
```

<p>This gives <code>C2=3/2</code>. Again we substitute in to get our answer:</p>

```julia
ex3 = subs(ex2, Sym("C2"), 3//2)
```

<h6>Example</h6>

<p>We do one more example, this one borrowed from <a href="http://nbviewer.ipython.org/github/garth-wells/IA-maths-Ipython/blob/master/notebooks/Lecture1.ipynb">here</a>.</p>

<blockquote><p>Find the variation of speed with time of a parachutist subject to a drag force of $k\cdot v^2$. </p></blockquote>

<p>The equation is</p>

<p>$$ \frac{m}{k} \frac{dv}{dt} = \alpha^2 - v^2. $$</p>

<p>We proceed through:</p>

```julia
t, m,k,alpha = symbols("t,m,k,alpha")
v = symbols("v", cls=SymFunction)
ex = Eq( (m/k)*diff(v(t),t), alpha^2 - v(t)^2 )
```

<p>We can "classify" this ODE with the method <code>classify_ode</code>. As this is not exported, we call it using indexing:</p>

```julia
ex[:classify_ode]()
```

<p>It is linear, but not solvable. Proceeding with <code>dsolve</code> gives:</p>

```julia
dsolve(ex, v(t))
```
